```python
import pandas as pd
```

```python
vine_table = pd.read_csv('vine_table.csv')
```

```python
vine_table.head()
```

```python
vine_df = vine_table[vine_table['total_votes'] >= 20]
vine_df.head()
```

```python
helpful_df = vine_df[(vine_df['helpful_votes'] / vine_df['total_votes'])>=0.5]
```

```python
vine_y = helpful_df[helpful_df['vine'] == 'Y']
vine_y.head()
```

```python
vine_n = helpful_df[helpful_df['vine'] == 'N']
vine_n.head()
```

```python
vine_y_counts = vine_y.count()[0]
vine_y_5star_counts = vine_y[vine_y['star_rating'] == 5].count()[0]
vine_y_5star_percenate = round(vine_y_5star_counts / vine_y_counts * 100, 2)
print("Total vine_y riviews:    ", vine_y_counts)
print("five star riviews:       ", vine_y_5star_counts)
print("Percentage of five start:", vine_y_5star_percenate)
```

```python
vine_n_counts = vine_n.count()[0]
vine_n_5star_counts = vine_n[vine_n['star_rating'] == 5].count()[0]
vine_n_5star_percenate = round(vine_n_5star_counts / vine_n_counts * 100, 2)
vine_n_5star_percenate
print("Total vine_n riviews:    ", vine_n_counts)
print("five star riviews:       ", vine_n_5star_counts)
print("Percentage of five start:", vine_n_5star_percenate)
```
